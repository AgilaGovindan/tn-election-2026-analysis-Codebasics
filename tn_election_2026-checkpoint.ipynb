```python
# Cell 1 — Import libraries and load data
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import os

print("Current folder:", os.getcwd())

# Load data
df21 = pd.read_csv('Data/tn_2021_results.csv')
df26 = pd.read_csv('Data/tn_2026_results new.csv')
master = pd.read_csv('Data/constituency_master.csv')

print("2021 data:", df21.shape)
print("2026 data:", df26.shape)
print("Master:", master.shape)
print("\n✅ Data loaded successfully!")
```

```python
# Cell 2 — Explore the data
print("=== 2021 COLUMNS ===")
print(df21.columns.tolist())

print("\n=== 2026 COLUMNS ===")
print(df26.columns.tolist())

print("\n=== MASTER COLUMNS ===")
print(master.columns.tolist())

print("\n=== 2021 SAMPLE ===")
print(df21.head(3))

print("\n=== 2026 SAMPLE ===")
print(df26.head(3))

print("\n=== MASTER SAMPLE ===")
print(master.head(3))
```

```python
# Cell 3 — Extract winners from each election
# Winner = candidate with highest votes in each constituency

winners21 = df21.loc[df21.groupby('ac_number')['votes'].idxmax()].copy()
winners26 = df26.loc[df26.groupby('ac_number')['votes'].idxmax()].copy()

print("=== 2021 WINNERS — Seats per party ===")
print(winners21['party'].value_counts().head(10))

print("\n=== 2026 WINNERS — Seats per party ===")
print(winners26['party'].value_counts().head(10))

print("\n=== Total constituencies check ===")
print("2021 winners:", len(winners21), "| 2026 winners:", len(winners26))
```

```python
# Cell 4 — Flip Analysis (Q2)
# Merge 2021 and 2026 winners on ac_number

merged = winners21[['ac_number','constituency','party','region']].merge(
    winners26[['ac_number','party']],
    on='ac_number',
    suffixes=('_21','_26')
)

# Identify flips
merged['flipped'] = merged['party_21'] != merged['party_26']

total_flips = merged['flipped'].sum()
no_flips = (~merged['flipped']).sum()

print(f"Total seats that FLIPPED: {total_flips}")
print(f"Total seats that did NOT flip: {no_flips}")
print(f"Flip percentage: {round(total_flips/234*100, 1)}%")

print("\n=== WHERE DID FLIPPED SEATS GO? ===")
flips = merged[merged['flipped']==True]
print(flips.groupby(['party_21','party_26']).size().sort_values(ascending=False).head(15))
```

```python
# Cell 5 — Bar chart: Seats won 2021 vs 2026
import matplotlib.pyplot as plt
import numpy as np

parties = ['TVK', 'DMK', 'AIADMK', 'INC', 'PMK', 'Others']
seats_21 = [0, 133, 66, 18, 5, 12]
seats_26 = [108, 59, 47, 5, 4, 11]

x = np.arange(len(parties))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, seats_21, width, label='2021', color='#2196F3', alpha=0.85)
bars2 = ax.bar(x + width/2, seats_26, width, label='2026', color='#FF5722', alpha=0.85)

ax.set_xlabel('Party', fontsize=12)
ax.set_ylabel('Seats Won', fontsize=12)
ax.set_title('Tamil Nadu Assembly Elections: Seats Won by Party\n2021 vs 2026', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(parties)
ax.legend(fontsize=11)
ax.set_ylim(0, 160)

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=10)

for bar in bars2:
    height = bar.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('chart1_seats_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart 1 saved!")
```

```python
# Cell 6 — Vote Share Analysis (Q3)

# Total votes per party per election
total_votes_21 = df21['votes'].sum()
total_votes_26 = df26['votes'].sum()

party_votes_21 = df21.groupby('party')['votes'].sum().reset_index()
party_votes_21['vote_share'] = round(party_votes_21['votes'] / total_votes_21 * 100, 2)
party_votes_21 = party_votes_21.sort_values('vote_share', ascending=False)

party_votes_26 = df26.groupby('party')['votes'].sum().reset_index()
party_votes_26['vote_share'] = round(party_votes_26['votes'] / total_votes_26 * 100, 2)
party_votes_26 = party_votes_26.sort_values('vote_share', ascending=False)

print("=== 2021 VOTE SHARE (Top 8) ===")
print(party_votes_21[['party','vote_share']].head(8).to_string(index=False))

print("\n=== 2026 VOTE SHARE (Top 8) ===")
print(party_votes_26[['party','vote_share']].head(8).to_string(index=False))
```

```python
# Cell 8 — Margin Analysis (Q6)

# Calculate total votes per constituency in 2026
total_votes_per_const = df26.groupby('ac_number')['votes'].sum().reset_index()
total_votes_per_const.columns = ['ac_number', 'total_votes']

# Merge with winners
winners26_margin = winners26.merge(total_votes_per_const, on='ac_number')

# Calculate winner vote share %
winners26_margin['win_vote_share'] = round(
    winners26_margin['votes'] / winners26_margin['total_votes'] * 100, 2
)

# Categorise margins
def margin_category(share):
    if share >= 50:
        return 'Dominant (50%+)'
    elif share >= 40:
        return 'Comfortable (40-50%)'
    elif share >= 35:
        return 'Moderate (35-40%)'
    else:
        return 'Narrow (Below 35%)'

winners26_margin['margin_cat'] = winners26_margin['win_vote_share'].apply(margin_category)

print("=== MARGIN CATEGORIES — 2026 ===")
print(winners26_margin['margin_cat'].value_counts())

print("\n=== AVERAGE WINNER VOTE SHARE ===")
print(round(winners26_margin['win_vote_share'].mean(), 2), "%")

print("\n=== NARROWEST WINS (Bottom 10) ===")
print(winners26_margin[['constituency','party','win_vote_share']]\
      .sort_values('win_vote_share').head(10).to_string(index=False))

print("\n=== DOMINANT WINS (Top 10) ===")
print(winners26_margin[['constituency','party','win_vote_share']]\
      .sort_values('win_vote_share', ascending=False).head(10).to_string(index=False))
```

```python
# Cell 9 — Margin Distribution Chart (Q6)

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Data
categories = ['Dominant\n(50%+)', 'Comfortable\n(40-50%)', 
              'Moderate\n(35-40%)', 'Narrow\n(Below 35%)']
counts = [13, 71, 86, 64]
colors = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Left — Bar chart
bars = ax1.bar(categories, counts, color=colors, alpha=0.88, edgecolor='white', linewidth=1.5)
ax1.set_title('How Dominant Were the Wins?\n2026 Tamil Nadu Election', 
              fontsize=13, fontweight='bold')
ax1.set_ylabel('Number of Constituencies', fontsize=11)
ax1.set_ylim(0, 110)

for bar, count in zip(bars, counts):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             str(count), ha='center', fontsize=12, fontweight='bold')

# Add total annotation
ax1.text(0.98, 0.95, 'Total: 234 seats', transform=ax1.transAxes,
         ha='right', fontsize=9, color='gray')

# Right — Pie chart
wedges, texts, autotexts = ax2.pie(
    counts, labels=categories, colors=colors,
    autopct='%1.1f%%', startangle=90,
    pctdistance=0.75, labeldistance=1.1,
    textprops={'fontsize': 9},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
for autotext in autotexts:
    autotext.set_fontweight('bold')

ax2.set_title('Share of Each Margin Category\n2026 Tamil Nadu Election',
              fontsize=13, fontweight='bold')

# Key insight text
fig.text(0.5, 0.01,
         '64 seats (27.4%) won with less than 35% vote share — a fragmented mandate',
         ha='center', fontsize=11, color='#e74c3c', fontweight='bold')

plt.tight_layout()
plt.subplots_adjust(bottom=0.1)
plt.savefig('chart3_margins.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart 3 saved!")
```
